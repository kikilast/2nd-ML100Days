# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):
    return abs(x)*(x>0)

def dReLU(x):
    a = 1
    return x*(x>=0)+a*(np.exp(x)-1)*(x<0)

x = plt.linspace(-10,10,100)
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,10)')
plt.grid()
plt.title('ReLU Function')
plt.text(4, 0.7, r'$f(x)=max(0,x)$', fontsize=15)
plt.show()
```

```python
x = plt.linspace(-10,10,100)

plt.plot(x, dReLU(x), 'r', label='linspace(-10,10,10)')
plt.grid()
plt.title('dReLU Function')
plt.text(4, 0.6, r'$f(x)=\{\stackrel{x, x\geq0}{\alpha(e^{x}-1), x<0}$', fontsize=15)
plt.show()
```
